````python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import seaborn as sns
import torch
from torch import nn

from torch.utils.data import DataLoader
from torch.utils.data import Dataset

import warnings
````

````python
points, labels = make_moons(n_samples=500, shuffle=True, noise=0.1, random_state=0)

sns.scatterplot(
    x=points[:, 0], y=points[:, 1], hue=labels, marker="o", s=25, edgecolor="k", legend=True
).set_title("Data")
plt.show()
````

````python
def input_validator(layers: list[int], activations: list[nn.Module]):
        if len(layers) - 1 != len(activations):
            warnings.warn("Number of activations doesn't correspond to number of layers.")

        # TODO warning if activation func not among activations in torch.nn
````

## MLP

````python
class MLP(nn.Module):
    def __init__(self, layer_sizes: list[int], activations: list[nn.Module]):
        input_validator(layer_sizes, activations)   
        super().__init__()
        self.num_layers = len(layer_sizes) - 1
        self.activations = nn.ModuleList(activations)
        self.layers = nn.ModuleList([
            nn.Linear(layer_sizes[i], layer_sizes[i+1]) for i in range(self.num_layers)
        ])
    
    def forward(self, x):
        for i in range(self.num_layers):
            x = self.layers[i](x)
            if (i < len(self.activations)):
                x = self.activations[i](x)
                
        if (self.num_layers > len(self.activations)):
            x = nn.Sigmoid()(x)
        return x
````

## Load data

````python
class MyCustomDataset(Dataset):
    def __init__(self, points, labels):
        self.X = torch.tensor(points).float()
        self.y = torch.tensor(labels).reshape(-1, 1).float()
        
    def __len__(self):
        assert(len(self.X) == len(self.y))
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
````

````python
MyDataset = MyCustomDataset(points, labels)
````

````python
def test(
    layers: list[int],
    activations: list[nn.Module],
    epochs=100,
    lr=1e-3,
    data = MyDataset,
    output=False,
    plot=True,
    debug=False,
):
    model = MLP(layers, activations)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.BCELoss()

    train_losses = []

    for epoch in range(1, epochs+1):
        if (output):
            print(f'{epoch=}')
        optimizer.zero_grad()
        outputs = model(data.X)
        
        if debug:
            print(f'{outputs.shape=}')
            print(f'{data.y.shape=}')
            print(data.y[:5])
            print(outputs[:5])
        
        
        loss = criterion(outputs, data.y)
        loss.backward()
        optimizer.step()
        train_losses.append(loss.item())
        
    if (plot):
        plt.plot(train_losses)
        plt.title('default testing')
        plt.show()
````

````python
def testDataloader(
    layers: list[int],
    activations: list[nn.Module],
    epochs=100,
    lr=1e-3,
    data = MyDataset,
    batch_size=32,
    output=False,
    plot=True,
    debug=False,
):
    model = MLP(layers, activations)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.BCELoss()

    train_losses = []

    dataloader = DataLoader(data, batch_size=batch_size, shuffle=True)

    for epoch in range(1, epochs+1):
        if output:
            print(f'{epoch=}')
        
        epoch_loss = 0
        for batch in dataloader:
            optimizer.zero_grad()
            batch_X, batch_y = batch
            outputs = model(batch_X)
            
            if debug:
                print(f'{outputs.shape=}')
                print(f'{batch_y.shape=}')
                print(batch_y[:5])
                print(outputs[:5])
            
            loss = criterion(outputs, batch_y)
            loss.backward()
            optimizer.step()
            
            epoch_loss += loss.item()
        
        train_losses.append(epoch_loss / len(dataloader))
        
    if (plot):
        plt.plot(train_losses)
        plt.title('dataloader testing')
        plt.show()
````

````python
activations = [nn.ReLU(), nn.Sigmoid()]
layers = [2, 5, 1]
test(
    layers=layers, 
    activations=activations,
    epochs=1000, 
    output=False
    )
testDataloader(
    layers=layers, 
    activations=activations,
    epochs=1000, 
    output=False
    )
````

````python
layers = [2, 20, 20, 20, 20, 1]
activations = [nn.ReLU(), nn.ReLU(), nn.ReLU(), nn.ReLU(), nn.Sigmoid()]
test(
    layers=layers, 
    activations=activations,
    epochs=500, 
    output=False
    )

testDataloader(
    layers=layers, 
    activations=activations,
    epochs=500, 
    output=False
    )
````

````python
layers = [2, 50, 50, 1]
activations = [nn.Hardswish(), nn.GELU(), nn.Sigmoid()]
test(
    layers=layers, 
    activations=activations,
    epochs=500, 
    output=False
    )
testDataloader(
    layers=layers, 
    activations=activations,
    epochs=500, 
    output=False
    )
````

````python
layers = [2, 50, 50, 1]
activations = [nn.Hardswish(), nn.GELU()]
test(
    layers=layers, 
    activations=activations,
    epochs=500, 
    output=False
    )
testDataloader(
    layers=layers, 
    activations=activations,
    epochs=500, 
    output=False
    )
````
